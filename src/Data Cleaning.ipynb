### Functions to decompose columns

This functions were created to perform similar actions to 'pd.get_dummies()'.Why don't we just use 'pd.get_dummies()'? The reason is that 1 row(movie) can have several genres/languages/countries and if we use 'pd.get_dummies()' it will create a column like 'Horror/Thriller' and set 1 to this column. We would like to avoid that and instead create 2 columns 'Horror' and 'Thriller' and set both of them to 1, instead.


```python
def add_to_set_from_series(set_,series_):
    for i in series_:
        tmp = str(i).split(',')
        for j in tmp:
            set_.add(j)
```

```python
def get_clean_set(set_):
    set_ = list(set_)
    for i in range(len(set_)):
        set_[i] = set_[i].strip()
    if 'nan' in set_:
        set_.remove('nan')
    return set(set_)
```

```python
def get_columns_with_values_from_set(set_,df, column_to_decompose):
    set_ = list(set_)
    for i in set_:
        df[column_to_decompose + '_' + i] = 0
    for i in set_:
        df.loc[df[column_to_decompose].apply(lambda x: i in x), column_to_decompose + '_' + i] += 1
    set_ = set(set_)
```

We think that directors/writers/actors/production companies play very important role in movie creation process. Though this information is represented as a string of people/companies who worked on this movie, which is not very useful for us. However, we wanted somehow to capture this information in our analysis:
1. We extracted each unique director/writer/actor/production company and calculated the number of movies they participated at.
2. Than we shortened this lists to top 10%.
3. For each movie, we calculated how many directors/writers/actors/production companies out of top 10% have been working on this movie.
4. This information is captured in 'n_top_...' columns.

<div class="alert alert-block alert-danger">Do <b>NOT</b> run this code. It might take you a long time to come back. For your convenience <b>output</b> of this code will be presented as a <b>separate file</b></div>

### Top 10% directors

```python
directors = set()
```

```python
add_to_set_from_series(directors, df.director)
```

```python
directors = get_clean_set(directors)
```

```python
directors = {i: 0 for i in directors}
```

```python
del directors['Unknown']
```

```python
for row in df.itertuples():
    row_direc = set()
    tmp = str(row.director).split(',')
    for j in tmp:
        row_direc.add(j)
    row_direc = list(row_direc)
    for i in range(len(row_direc)):
        row_direc[i] = row_direc[i].strip()
    if 'nan' in row_direc:
        row_direc.remove('nan')
    row_direc = set(row_direc)
    for key, value in directors.items():
        if key in row_direc:
            directors[key] += 1
```

```python
directors = pd.Series(directors)
```

```python
directors.describe(percentiles = [0.90])
```

```python
top_directors = directors[directors > 5].index
```

```python
df['n_top_directors'] = 0
```

```python
for row in df.itertuples():
    row_direc = set()
    tmp = str(row.director).split(',')
    for j in tmp:
        row_direc.add(j)
    row_direc = list(row_direc)
    for i in range(len(row_direc)):
        row_direc[i] = row_direc[i].strip()
    if 'nan' in row_direc:
        row_direc.remove('nan')
    row_direc = set(row_direc)
    for i in row_direc:
        if i in top_directors:
            df.loc[df.imdb_title_id == row.imdb_title_id, 'n_top_directors'] += 1
```
